```python
df = pd.read_csv('Google-Playstore.csv')
df.head()
```

## Preliminary Wrangling
> ## Data cleaning

```python
df.info()
```

### Drop nan value

```python
df.Size = df.Size.replace('Varies with device',np.nan)
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
```

### Remove space in colmuns name

```python
df.columns = df.columns.str.replace(' ', '') 
```

### Convert string data types to float in [Size, Installs]

```python
df.Size = df.Size.str.replace('k','e+3')
df.Size = df.Size.str.replace('M','e+6')
df.Size = df.Size.str.replace('G','e+9')
df.Size = df.Size.str.replace(',', '') 
df.Size = df.Size.astype(float)
df.Size = df.Size / 1024 
df.Size.head()
```

```python
df.Installs = df.Installs.apply(lambda x: x.strip('+'))
df.Installs = df.Installs.replace('Free',np.nan)
df.Installs = df.Installs.str.replace(',','')
df.Installs = df.Installs.astype(float)
df.Installs.head()
```

## Univariate Exploration

> In this section, investigate distributions of individual variables. If
you see unusual points or outliers, take a deeper look to clean things up
and prepare yourself to look at relationships between variables.

```python
df_ = df_ = df.query("RatingCount > 5000")
def bar_plot(title, data, fig_size, rotation=0):
    plt.title(title)
    base_color = sb.color_palette()[0]
    fig = plt.gcf()
    fig.set_size_inches(fig_size[0], fig_size[1])
    sb.countplot(data=df, x=data, color=base_color, order = df[data].value_counts().index);
    plt.xticks(rotation=rotation);
    
def hist_plot(title, xlabel, ylabel, data, step, rotation=0):
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xticks(np.arange(0, df[data].max()+1, step=step))
    bins = np.arange(0, df[data].max()+1, step)
    plt.hist(data = df_, x = data, bins = bins);
    plt.xticks(rotation=rotation);

def bar_plot2d(title, dataframe, x_, y_, rotation=0):
    base_color = sb.color_palette()[0]
    plt.figure(figsize=(18, 10))
    plt.title(title)
    res = dataframe.groupby(x_)[y_].mean().reset_index()
    finaldata = res.sort_values(y_, ascending=False)
    sb.barplot(data=finaldata, x=x_, y=y_, order=np.array(finaldata[x_]), color=base_color);
    plt.xticks(rotation=rotation);
    finaldata[:5]
```

## Which apps Category has the highest rating? (with rating count > 5000)

```python
base_color = sb.color_palette()[0]
plt.figure(figsize=(18, 8))
df_ = df.query("RatingCount > 5000")
res = df_.groupby('Category')['Rating'].mean().reset_index()
finaldata = res.sort_values('Rating', ascending=False)
sb.barplot(data=finaldata, x='Category', y='Rating', order=np.array(finaldata.Category), color=base_color);
plt.xticks(rotation=90);
finaldata[:5]
```

> The bar plot shows the relationship between app category and the rating.  
NOTE: the plot may not be very accurate because the rating count varies from app to app (here I choose to plot only the apps with ratings above 5000 counts).  
> The top 5 apps category in number of installs:  
1- Books & Reference.  
2- Parenting.  
3- Health & Fitness.  
4- Word.  
5- Casino.

## Which apps Category has the most number of installs?

```python
base_color = sb.color_palette()[0]
plt.figure(figsize=(18, 8))
res = df.groupby('Category')['Installs'].mean().reset_index()
finaldata = res.sort_values('Installs', ascending=False)
sb.barplot(data=finaldata, x='Category', y='Installs', order=np.array(finaldata.Category), color=base_color);
plt.xticks(rotation=90);
finaldata[:5]
```

> The bar plot shows the relationship between app category and the number of installs of the app.  
> The top 5 apps category in number of installs:  
1- Racing.  
2- Action.  
3- Strategy.  
4- Arcade.  
5- Video Players & Editors.
